```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
table1 = pd.read_excel('coefficients.xlsx', sheet_name=2, header=0)
table2 = pd.read_excel('coefficients.xlsx', sheet_name=1, header=0)
```

```python
model = Sequential()
model.add(Dense(2, input_dim=2, activation='sigmoid'))
model.add(Dense(4, activation='sigmoid'))
model.add(Dense(3, activation='linear'))

model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
```

```python
x = table1[['z', 'L0/H']]
y = table1.drop(['z', 'L0/H'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=3, test_size=0.1)
```

```python
model.fit(x_train, y_train, epochs=300, batch_size=None)

scores = model.evaluate(x_test, y_test)
scores[1]
```

```python
predictions = model.predict(table2)
print(predictions)
```
