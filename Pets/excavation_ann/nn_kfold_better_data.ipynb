```python
import numpy as np
import pandas as pd

from keras.models import Sequential
from keras.layers import Dense
```

```python
table1 = pd.read_excel('coefficients.xlsx', sheet_name=4, header=0)
table2 = pd.read_excel('coefficients.xlsx', sheet_name=5, header=0)
```

```python
x = table1[['z', 'L0/H']]
y = table1.drop(['z', 'L0/H'], axis=1)

x = x.to_numpy(dtype='float32')
y = y.to_numpy(dtype='float32')
```

```python
model = Sequential()
model.add(Dense(2, input_dim=2, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(3, activation='linear'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10, shuffle=True, random_state=3)
```

```python
predictions_list = []
scores_list = []

for train, test in kf.split(x, y):
    x_train, x_test, y_train, y_test = x[train], x[test], y[train], y[test]

    model.fit(x_train, y_train, epochs=300, batch_size=None)
    scores_list.append(model.evaluate(x_test, y_test)[1])
    predictions_list.append(model.predict(table2))

print(np.mean(scores_list), np.mean(predictions_list, axis=0), sep='\n')
```
