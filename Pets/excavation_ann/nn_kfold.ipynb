```python
import numpy as np
import pandas as pd

from keras.models import Sequential
from keras.layers import Dense
```

```python
table1 = pd.read_excel('coefficients.xlsx', sheet_name=0, header=0)
table2 = pd.read_excel('coefficients.xlsx', sheet_name=1, header=0)
```

```python
x = table1[['z', 'L0/H']]
y = table1.drop(['z', 'L0/H'], axis=1)
```

```python
x = x.to_numpy(dtype='float32') #Преобразуем Pandas DataFrame в NumPy arrray
y = y.to_numpy(dtype='float32')
x #Выводим конвертированную таблицу для проверки
```

```python
model = Sequential()
model.add(Dense(2, input_dim=2, activation='sigmoid'))
model.add(Dense(4, activation='sigmoid'))
model.add(Dense(3, activation='linear'))

model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
```

```python
from sklearn.model_selection import KFold #Импортируем метод KFold
kf = KFold(n_splits=10, shuffle=True, random_state=3) #Задаём кол-во валидаций и размер тестовой выборки
```

```python
predictions_list = [] #Создаём пустой список для записи результатов по не встречавшимся значениям
scores_list = [] #Пустой список для значений точности

for train, test in kf.split(x, y): #Цикл: для каждого значения train и test из объекта kf.split
    x_train, x_test, y_train, y_test = x[train], x[test], y[train], y[test] #Инициализируем выборки

    model.fit(x_train, y_train, epochs=300, batch_size=None) #Тренируем модель
    scores_list.append(model.evaluate(x_test, y_test)[1]) #Записываем значение точности в список
    predictions_list.append(model.predict(table2)) #Записываем в список результаты по уникальным данным

print(np.mean(scores_list), np.mean(predictions_list, axis=1), sep='\n')
#Выводим средние значения точности и результатов после цикла
```
