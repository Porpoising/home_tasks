```python
import pandas as pd
import seaborn as sb
```

```python
students_performance = pd.read_csv('D:\Home tasks\Datasets\StudentsPerformance.csv')
students_performance.columns = [x.replace(" ", "_") for x in students_performance.columns]
students_performance
```

```python
# строим гистограмму колонки math_score (метод matplotlib)
students_performance.math_score.hist()
```

```python
# строим scatter-plot колонок math_score и reading_score (метод matplotlib)
students_performance.plot.scatter(x='math_score', y='reading_score')
```

```python
# scatter-plot в seaborn
ax = sb.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg=False)
# hue - группирующая переменная, fit_reg=False - бе регрессионных прямых
ax.set_xlabels('Math Score')
ax.set_ylabels('Reading Score')
```

```python
# heatmap - график
genome = pd.read_csv('D:\Home tasks\Datasets\genome_matrix.csv', index_col=0) # index_col - принять столбец в качестве индексов
genome_plot = sb.heatmap(data=genome, cmap='viridis') # cmap - выбор цветовой схемы
genome_plot.xaxis.set_ticks_position('top') # надписи сверху
genome_plot.xaxis.set_tick_params(rotation=90) # поворот надписей на 90 гр. по часовой
```
