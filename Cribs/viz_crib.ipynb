# Plotly

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)
```

## Line plot

```python
trace0= go.Scatter(x=students_performance['reading_score'],
                    y=students_performance['math_score'], name='Read/Math line', mode='markers')

trace1 = go.Scatter(x=students_performance['reading_score'],
                    y=students_performance['writing_score'], name='Read/Write line', mode='markers')

data = [trace0, trace1]
layout = {'title': 'Read lines'}

fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

## Bar plot

```python
lg = pd.crosstab(students_performance['lunch'], students_performance['gender']).reset_index()
lg
```

```python
fig = go.Figure(data=[
    go.Bar(name='Female lunch', x=lg['lunch'], y=lg['female']),
    go.Bar(name='Male lunch', x=lg['lunch'], y=lg['male'])
])

fig.update_layout(barmode='group')
fig.show()
```

## Box plot

```python
fig = go.Figure()
fig.add_trace(go.Box(y=students_performance['math_score'], quartilemethod="linear", name="Linear Quartile Mode"))
fig.add_trace(go.Box(y=students_performance['math_score'], quartilemethod="inclusive", name="Inclusive Quartile Mode"))
fig.add_trace(go.Box(y=students_performance['math_score'], quartilemethod="exclusive", name="Exclusive Quartile Mode"))
fig.update_traces(boxpoints='all', jitter=0)
fig.show()
```
