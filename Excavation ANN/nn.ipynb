```python
import pandas as pd #Подключаем библиотеку Pandas и даём ей имя объекта pd
from keras.models import Sequential #Подключаем класс Sequential из библиотеки Keras
from keras.layers import Dense #Подключаем класс Dense из библиотеки Keras
```

```python
table1 = pd.read_excel('coefficients.xlsx', sheet_name=0, header=0) #Считываем данные таблицы
table2 = pd.read_excel('coefficients.xlsx', sheet_name=1, header=0) #и передаём в table1 и table2
```

```python
table1.tail() #Выводим последние пять строк таблицы
```

```python
model = Sequential() #Создаём объект model класса Sequential
model.add(Dense(2, input_dim=2, activation='sigmoid'))
model.add(Dense(4, activation='sigmoid'))
model.add(Dense(3, activation='linear'))

model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
```

```python
x = table1[['z', 'L0/H']] #Присваиваем входные столбцы объекту х
y = table1.drop(['z', 'L0/H'], axis=1) #Присваиваем копию первой таблицы без входных столбцов
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split #Импортируем метод из библиотеки
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=3, test_size=0.1) #Разделяем
```

```python
model.fit(x_train, y_train, epochs=300, batch_size=None) #Подаём данные и тренируем сеть

scores = model.evaluate(x_test, y_test) #Проверяем модель на тестовых данных
scores[1]
```

```python
predictions = model.predict(table2) #Даём сети абсолютно новые данные
print(predictions) #Выводим новые результаты
```
